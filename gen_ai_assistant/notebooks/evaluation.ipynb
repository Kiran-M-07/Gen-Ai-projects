# Evaluating details extraction

```python
import pandas as pd 
from sentence_transformers import SentenceTransformer, util
```

```python
model = SentenceTransformer('all-mpnet-base-v2')
```

```python
def cos_sim(model,x1,x2):
    
    embedding1 = model.encode(str(x1), convert_to_tensor=True)
    embedding2 = model.encode(str(x2), convert_to_tensor=True)
    similarity = util.cos_sim(embedding1,embedding2)
    return round(similarity.item(),2)
    
```

```python
df_insights_meta = pd.read_csv("../outputs/insights_meta.csv")
```

```python
df_insights_meta["summ_abstract_sim"] = df_insights_meta.apply(lambda row: cos_sim(model,row["summary"],row["abstract"]), axis = 1)
```

```python
df_insights_meta["summ_abstract_sim"].min(),df_insights_meta["summ_abstract_sim"].max(), df_insights_meta["summ_abstract_sim"].mean()
```

```python
df_insights_meta.to_csv("../outputs/insights_meta.csv", index = False)
```
